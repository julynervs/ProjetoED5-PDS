```python
%matplotlib inline
import io
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack
import urllib.request
```

```python
import cv2

url='http://i.imgur.com/8vuLtqi.png'
url_response = urllib.request.urlopen(url)
img_array = np.array(bytearray(url_response.read()), dtype=np.uint8)
img = cv2.imdecode(img_array, cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap=plt.cm.gray)
plt.title('Imagem\n');
```

```python
url2='https://images.vexels.com/media/users/3/143600/isolated/preview/1b2a1e1747f67ce87ea8af5fdf410d23-yao-ming-enfrenta-o-meme-by-vexels.png'
url_response2 = urllib.request.urlopen(url2)
img_array2 = np.array(bytearray(url_response2.read()), dtype=np.uint8)
img2 = cv2.imdecode(img_array2, cv2.IMREAD_GRAYSCALE)
res = cv2.resize(img2,None,fx=0.4, fy=0.4, interpolation = cv2.INTER_CUBIC)

plt.imshow(res, cmap=plt.cm.gray)
plt.title('Marca dágua\n');
```

```python
dct = fftpack.dct(fftpack.dct(img.T, norm='ortho').T, norm='ortho')

plt.matshow(np.log(np.abs(dct)), cmap=plt.cm.Paired);
plt.title('DCT da Imagem\n');
```

```python
[l1, c1] = img.shape
[l2, c2] = res.shape


dct[(l1-l2+1):l1,(c1-c2+1):c1] = res[1:l2,1:c2]


plt.imshow(dct, cmap=plt.cm.gray)
plt.title('Marca dágua na DCT\n');
```
